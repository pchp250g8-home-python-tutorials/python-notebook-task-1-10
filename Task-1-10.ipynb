При каком натуральном значении n дробь (2 * n ** 3 - 2 * n ** 2 - 11 * n + 3) / (n + 1) является натуральным чилом? 
В ответ записать сумму этих значний n.

```python
from sympy import pdiv
from sympy.abc import n
from sympy import divisors
from sympy import apart
from numpy import sum
from sympy import solve
from sympy import Eq

print(pdiv(2 * n ** 3 - 2 * n ** 2 - 11 * n + 3, n + 1))
print(apart((2 * n ** 3 - 2 * n ** 2 - 11 * n + 3) / (n + 1)))
d = divisors(10); c = 0; s = 0
for i in d:
    eq1 = Eq(10 / (n + 1), i)
    sol = solve(eq1,n)
    print("Equation N",(c+1),eq1.lhs,"=",eq1.rhs,"Solution:",sol)
    m = sol.pop()
    if(m > 0):
        s += m
    c += 1
print("The sum of numbers is:",s)
```
